```python
# Filtrer le DataFrame
df_volailles = df_dispo_alimentaire[
    (df_dispo_alimentaire['Produit'] == 'Viande de Volailles') &
    (df_dispo_alimentaire['Élément'].isin([
        'Production',
        'Importations - Quantité',
        'Disponibilité intérieure',
        'Nourriture'
    ]))
]
```

```python
df_volailles.head()
```

```python
# 1. Liste de tous les pays uniques
pays_uniques = pd.DataFrame({'Zone': df_dispo_alimentaire['Zone'].unique()})

# 2. Pivot sur la viande de volailles avec les dimensions choisies
pivot_volailles = df_volailles.pivot_table(
    index=['Zone',  'Code Produit', 'Code zone', 'Produit', 'Année','Unité'],
    columns='Élément',
    values='Valeur',
    aggfunc='sum'
).reset_index()

# 3. Merge pour inclure tous les pays, même ceux sans données
df_dispo_alimentaire_final = pays_uniques.merge(
    pivot_volailles,
    on='Zone',
    how='left'
)

```

## Mapping pays


```python
# Définir la locale française
locale_fr = Locale('fr')

# Lister les pays
pays_liste = []

for country in pyc.countries:
    try:
        nom_fr = locale_fr.territories.get(country.alpha_2)
        pays_liste.append({
            'Nom_FR': nom_fr,
            'Nom_EN': country.name,
            'Alpha_2': country.alpha_2,
            'Alpha_3': country.alpha_3,
            'Numérique': country.numeric
        })
    except KeyError:
        continue  # certains territoires n'ont pas de traduction
```

```python
df_mapping_en_pays = pd.DataFrame(pays_liste)
```

```python
df_mapping_en_pays.head()
```

```python
df_mapping_en_pays= df_mapping_en_pays.rename(columns={'Nom_FR':'Zone', 'Alpha_3' : 'Country Code'})
```

```python
colonne_mapping_en_pays_a_garder = ['Zone', 'Nom_EN', 'Country Code']
```

```python
df_mapping_en_pays = df_mapping_en_pays[colonne_mapping_en_pays_a_garder]
```

```python
df_mapping_en_pays['Zone'] = df_mapping_en_pays['Zone'].str.upper()
```

```python
df_mapping_en_pays.head(30)
```

```python
# 2. Charger le fichier CSV de correspondance
mapping_df = pd.read_csv("correspondance_pays.csv", encoding='latin1', sep = ';') # remplace par ton vrai nom de fichier

# 3. Créer un dictionnaire de correspondance à partir du mapping
mapping_dict = dict(zip(
    mapping_df['Zone de df_dispo_alimentaire_final'],
    mapping_df['Zone df_mapping_en_pays']
))

# 4. Appliquer la correction sur la colonne 'Zone' de df
df_dispo_alimentaire_final['Zone_corrigée'] = df_dispo_alimentaire_final['Zone'].map(mapping_dict)

# 5. Optionnel : conserver l'original si pas trouvé dans le mapping
df_dispo_alimentaire_final['Zone_corrigée'] = df_dispo_alimentaire_final['Zone_corrigée'].fillna(df_dispo_alimentaire_final['Zone'])

# 6. Afficher ou enregistrer le résultat
print(df_dispo_alimentaire_final[['Zone', 'Zone_corrigée']])
```

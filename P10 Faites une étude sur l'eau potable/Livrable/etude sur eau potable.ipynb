### Comparaison Pays

```python

# R√©cup√®re les sets de pays uniques √† partir de tes DataFrames
mortality_countries = set(pivot_mortality['Country'].unique())
water_countries = set(df_water_services_copy['Country'].unique())
political_countries = set(df_political['Country'].unique())
population_countries = set(df_population_copy['Country'].unique())
region_countries = set(df_region_copy['Country'].unique())

# Dictionnaire des tables
tables = {
    'pivot_mortality': mortality_countries,
    'water_services': water_countries,
    'political': political_countries,
    'population': population_countries,
    'region': region_countries,
}

# Boucle pour identifier les pays manquants
for table_a, countries_a in tables.items():
    for table_b, countries_b in tables.items():
        if table_a != table_b:
            missing = countries_a - countries_b
            if missing:
                print(f"üü• Pays pr√©sents dans {table_a} mais absents de {table_b} ({len(missing)} pays) :")
                print(sorted(missing))
                print()

## on garde :
##China 
##on renomme republic_of_north_macedonia
##on renomme sudan --> NON en r√©alit√© Soudan apr√®s 2012 = Soudan + Sudan du Sud
##exclure les zones non reconnues
```

```python
df_population_copy = df_population_copy[~df_population_copy['Country'].isin(['China'])] ##On va garder "China, mainland" donc on supprime chine qui est l'addition de mainland + les iles (Taiwan, Hong Kong, Macao)
```

```python
# Dictionnaire de correspondance des pays selon les standards de l'OMS
country_mapping = {
    'China': 'China',
    'China, mainland': 'China',
    'China, Hong Kong SAR': 'Hong Kong',
    'China, Macao SAR': 'Macao',
    'China, Taiwan Province of': 'Taiwan',
    'Republic of North Macedonia': 'North Macedonia',

}

# Fonction pour harmoniser les noms des pays
def harmonize_country_names(df, column='Country'):
    df = df.copy()
    df[column] = df[column].replace(country_mapping)
    return df

# Application de la fonction sur chaque table

# Harmoniser les noms de pays dans pivot_mortality
pivot_mortality = harmonize_country_names(pivot_mortality)

# Harmoniser les noms de pays dans df_water_services_copy
df_water_services_copy = harmonize_country_names(df_water_services_copy)

# Harmoniser les noms de pays dans df_political
df_political = harmonize_country_names(df_political)

# Harmoniser les noms de pays dans df_population_copy
df_population_copy = harmonize_country_names(df_population_copy)

# Harmoniser les noms de pays dans df_region
df_region_copy = harmonize_country_names(df_region_copy)
```

```python
# Exporter chaque table apr√®s harmonisation en format CSV

# Exporter pivot_mortality
pivot_mortality.to_csv('df_mortality_clean.csv', index=False)

# Exporter df_water_services_copy
df_water_services_copy.to_csv('df_water_services_clean.csv', index=False)

# Exporter df_political
df_political.to_csv('df_political_clean.csv', index=False)

# Exporter df_population_copy
df_population_copy.to_csv('df_population_clean.csv', index=False)

# Exporter df_region_copy
df_region_copy.to_csv('df_region_clean.csv', index=False)
```
